```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy
```

```python
#가우시안분포 활용한 랜덤 좌표 생성
#g0=np.random.multivariate_normal([1,1], np.eye(2), 100)
#g1=np.random.multivariate_normal([3,5], np.eye(2), 100)
#g2=np.random.multivariate_normal([9,9], np.eye(2), 100)
g0=np.random.rand(100, 2)
g1=np.random.rand(100, 2)
g2=np.random.rand(100,2)
x=np.vstack([g0, g1, g2])
x=np.asmatrix(x)
print(x)
plt.figure(figsize=(10,10))
plt.plot(x[:, 0], x[:, 1], 'b.')
plt.axis('equal')
plt.show()
```

```python
k=3
m=x.shape[0]

mu = x[np.random.randint(0, m, k), :]
pre_mu = copy.deepcopy(mu)
print(mu)
plt.figure(figsize=(10,10))
plt.plot(x[:, 0], x[:, 1], 'b.')
plt.plot(mu[:, 0], mu[:, 1], 's', color='r')
plt.axis('equal')
plt.show()
```

```python
y = np.empty([m,1])
for j in range(1000):    
    for i in range(m):
        d0=np.linalg.norm(x[i,:]-mu[0,:],2)
        d1=np.linalg.norm(x[i,:]-mu[1,:],2)
        d2=np.linalg.norm(x[i,:]-mu[2,:],2)
        y[i]=np.argmin([d0, d1, d2])
    for i in range(k):
        mu[i, :] = np.mean(x[np.where(y==i)[0]], axis = 0)
    print("current step is... ", j)
    print(mu)
```

```python
x0 = x[np.where(y==0)[0]]
x1=x[np.where(y==1)[0]]
x2=x[np.where(y==2)[0]]
plt.figure(figsize=(10,10))
plt.plot(x0[:,0], x0[:,1], 'r.', label='C0')
plt.plot(x1[:,0], x1[:,1], 'g.', label='C1')
plt.plot(x2[:,0], x2[:,1], 'b.', label='C2')
plt.plot(mu[:,0], mu[:, 1], 's', color='black')
plt.axis('equal')
plt.legend(fontsize = 12)
plt.show()
```
